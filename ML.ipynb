De las variables más importantes Renta es la que tiene más valores nulos por lo tanto ocuparemos el método chi-cuadrado para
ver si existe relación entre los nulos de esta columna con los demás datos faltantes.

```python
# Suponiendo que df es tu DataFrame y que 'Renta' es la columna principal
nulos_columnas = df.columns[df.isnull().any()].tolist()
resultados = {}

count_bajo_0_05 = 0
count_sobre_0_05 = 0

for col in nulos_columnas:
    contingency_table = pd.crosstab(df['Renta'].isnull(), df[col].isnull())
    chi2, p, _, _ = chi2_contingency(contingency_table)
    resultados[col] = p
    
    if p < 0.05:
        count_bajo_0_05 += 1
    else:
        count_sobre_0_05 += 1
    
    print(f'Valor p para la comparación con {col}: {p}')

print()
print(f'Número de valores p menores a 0.05: {count_bajo_0_05}')
print(f'Número de valores p mayores o iguales a 0.05: {count_sobre_0_05}')
```

Como podemos apreciar en el conteo la mayoría de variables tienen un valor p mayor a 0.05 por lo tanto podemos concluir que
en su mayoría no existe relación entre los nulos del dataframe. Por lo tanto, los datos faltantes son aleatorios.
Esta información nos será de ayuda más adelante

Ahora creamos una función que compare los 3 modelos y seleccione la mejor automáticamente según las métricas.

```python
def mejor_modelo_regresion(X_train, y_train, X_test, y_test):
    # Diccionario para almacenar los modelos y sus nombres
    modelos = {
        'Regresión Lineal Múltiple': LinearRegression(),
        'Árbol de Decisión': DecisionTreeRegressor(max_depth=5, min_samples_leaf=10),
        'SVR': LinearSVR()
    }
    
    # Diccionario para almacenar los resultados de las métricas
    resultados = {}

    # Entrenar y evaluar cada modelo
    for nombre, modelo in modelos.items():
        # Entrenar el modelo
        modelo.fit(X_train, y_train)
        
        # Realizar predicciones sobre los datos de prueba
        y_pred = modelo.predict(X_test)
        
        # Calcular las métricas
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        # Almacenar las métricas en el diccionario de resultados
        resultados[nombre] = {
            'MSE': mse,
            'MAE': mae,
            'R²': r2
        }

    # Determinar el mejor modelo según el MSE más bajo (puedes cambiar por otra métrica si prefieres)
    mejor_modelo = min(resultados, key=lambda x: resultados[x]['MSE'])
    
    # Imprimir los resultados
    print("Resultados:")
    for nombre, metricas in resultados.items():
        print(f"\nModelo: {nombre}")
        print(f"  MSE: {metricas['MSE']}")
        print(f"  MAE: {metricas['MAE']}")
        print(f"  R²: {metricas['R²']}")
    
    print(f"\nEl mejor modelo es: {mejor_modelo}")
    
    return mejor_modelo, modelos[mejor_modelo]
```

```python
mejor_modelo_regresion(X_train, y_train, X_test, y_test)
```

La función ha establecido que el mejor modelo de regresión es la regresión lineal múltiple,
eso se debe a que tiene un R² más alto lo que indica un mayor grado de ajuste del modelo a los datos,
además de tener un MSE y MAE más bajo lo que indica que comete menos errores a la hora de hacer las predicciones.
Todo esto en conjunto hace que este sea el mejor modelo.

```python
def mejor_modelo_clasificacion(X_train, y_train, X_test, y_test):
    # Diccionario para almacenar los modelos y sus nombres
    modelos = {
        'KNN': KNeighborsClassifier(n_neighbors=5),
        'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
        'SVM': SVC(kernel='linear', probability=True, random_state=42)
    }
    
    # Diccionario para almacenar los resultados de las métricas
    resultados = {}

    # Entrenar y evaluar cada modelo
    for nombre, modelo in modelos.items():
        # Entrenar el modelo
        modelo.fit(X_train, y_train)
        
        # Realizar predicciones sobre los datos de prueba
        y_pred = modelo.predict(X_test)
        
        # Calcular las métricas
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='weighted')
        recall = recall_score(y_test, y_pred, average='weighted')
        f1 = f1_score(y_test, y_pred, average='weighted')
        
        # Calcular la matriz de confusión
        cm = confusion_matrix(y_test, y_pred)
        # Falsos positivos (FP) y falsos negativos (FN) solo para problemas binarios
        fp, fn = 0, 0
        if cm.shape == (2, 2):  # Para problemas binarios
            tn, fp, fn, tp = cm.ravel()
        
        # Almacenar las métricas en el diccionario de resultados
        resultados[nombre] = {
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1-score': f1,
            'Falsos Positivos (FP)': fp,
            'Falsos Negativos (FN)': fn
        }

    # Determinar el mejor modelo según F1-score y luego por FP + FN si hay empate en F1-score
    mejor_modelo = min(resultados, key=lambda x: (-resultados[x]['F1-score'], resultados[x]['Falsos Positivos (FP)'] + resultados[x]['Falsos Negativos (FN)']))
    
    # Imprimir los resultados
    print("Resultados:")
    for nombre, metricas in resultados.items():
        print(f"\nModelo: {nombre}")
        print(f"  Accuracy: {metricas['Accuracy']:.2f}")
        print(f"  Precision: {metricas['Precision']:.2f}")
        print(f"  Recall: {metricas['Recall']:.2f}")
        print(f"  F1-score: {metricas['F1-score']:.2f}")
        if metricas['Falsos Positivos (FP)'] or metricas['Falsos Negativos (FN)']:
            print(f"  Falsos Positivos (FP): {metricas['Falsos Positivos (FP)']}")
            print(f"  Falsos Negativos (FN): {metricas['Falsos Negativos (FN)']}")
    
    print(f"\nEl mejor modelo es: {mejor_modelo}")
    
    return mejor_modelo, modelos[mejor_modelo]
```
